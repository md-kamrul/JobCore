```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Annotated
from langchain_core.messages import BaseMessage, HumanMessage
from langchain_openai import ChatOpenAI
```

**We are creating a chat state to manage the states of whole chat history. We are defining user message, system message, tool message and AI message on "BaseMessage" variable.**

**"add_messages" is a function here. In this context, add_messages is telling LangGraph how to handle updates to the messages key within the ChatState. When a new message is added to the state, add_messages ensures that the new message is appended to the existing list of messages, rather than replacing the entire list.**

```python
from langgraph.graph.message import add_messages

class ChatState(TypedDict):

    messages: Annotated[list[BaseMessage], add_messages]
```

```python
llm = ChatOpenAI()


def chat_node(state: ChatState):

    # take user query from state
    messages = state['messages']

    # send to llm
    response = llm.invoke(messages)

    # response store state
    return {'messages': [response]}
```

```python
graph = StateGraph(ChatState)

# add nodes
graph.add_node('chat_node', chat_node)

graph.add_edge(START, 'chat_node')
graph.add_edge('chat_node', END)

chatbot = graph.compile()
```

```python
initial_state = {
    'messages': [HumanMessage(content='What is the capital of india')]
}

chatbot.invoke(initial_state)['messages'][-1].content
```
